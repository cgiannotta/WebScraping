```python
import requests
from bs4 import BeautifulSoup
```

```python
# save url for request
URL = 'https://medium.com/better-programming/how-to-create-and-publish-a-chrome-extension-in-20-minutes-6dc8395d7153'

# scrape page
page = requests.get(URL)

# create BeautifulSoup object
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# take a list of urls, return dictionary of key information
def get_article_info(url_list):
    
    # list to store dictionaries of information for each article
    articles = []

    for url in url_list:
        # create dictionary to store article information
        article_dict = {'title': "", 'author': "", 'date': "", 'text': ""}
        
        # scrape page
        page = requests.get(url)
        
        # create BeautifulSoup object
        soup = BeautifulSoup(page.content, 'html.parser')
        
        # get key article information
        article_info = soup.find_all("title")[0].get_text().split(" | ")
        article_dict['title'] = article_info[0]
        article_dict['author'] = article_info[1].partition("by ")[2]
        article_dict['date'] = soup.find("meta",  property="article:published_time")["content"]
        
        # get article text
        paragraphs = list(soup.find_all('p'))
        article_text = []
        for paragraph in paragraphs:
            article_text.append(paragraph.get_text())
        sep = " "
        article_dict['text'] = sep.join(article_text)
        
        # add dictionary to articles list
        articles.append(article_dict)
        
    return articles
```

```python
urls = ['https://medium.com/better-programming/how-to-use-react-hooks-like-professional-573cc96fb79d', 'https://medium.com/better-programming/how-to-create-and-publish-a-chrome-extension-in-20-minutes-6dc8395d7153']
get_article_info(urls)
```
